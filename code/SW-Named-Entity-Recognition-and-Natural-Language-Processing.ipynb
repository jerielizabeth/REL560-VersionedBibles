```python
import nltk
import os
import pandas as pd
```

```python
source_dir = "../data/analysis_data/bibles-txt-sw/"

texts = []

for filename in os.listdir(source_dir):
    if filename.startswith('.'):
        print(filename)
        pass
    else:
        with open(os.path.join(source_dir, filename), 'r') as obit:
            content = obit.read()
        texts.append(
            {
                "doc_id": filename,
                "text": content
            }
        )
    
texts_df = pd.DataFrame(texts)

texts_df.head()
```

```python
texts_df.iloc[2]['text']
ethel = texts_df.iloc[2]['text']

ethel
```

```python
# Splitting into sentences on period
e_sentences = ethel.split('.')

e_sentences
```

```python
# Getting the tokens
e_parsed = []
for sentence in e_sentences:
    tokens = nltk.word_tokenize(sentence)
    e_parsed.append(tokens)
    
e_parsed
```

```python
# Figure out the parts of speech for the tokens it previously identified
e_tagged = []
for parsed_sentence in e_parsed:
    tagged = nltk.pos_tag(parsed_sentence)
    e_tagged.append(tagged)
    
e_tagged
```

```python
# Chunk entities together to name a unit
e_entities = []

for tagged_sent in e_tagged:
    entities = nltk.chunk.ne_chunk(tagged_sent)
    e_entities.append(entities)
    
e_entities
```
